# Simulated Streaming Example

```python
import numpy as np
import bokeh.models.sources 
import bokeh.plotting 
import bokeh.io
from bokeh.palettes import Category10_10 as palette
```

```python
bokeh.io.output_notebook()
```

```python
# # uncomment and run this cell if not connected to the internet
# import bokeh.resources
# bokeh.plotting.output_notebook(resources=bokeh.resources.INLINE)
```

```python
mu_x1, mu_y1 = np.random.randint(-40, 40, 2)
sigma_x1, sigma_y1 = np.random.randint(1, 25, 2)


my_fig = bokeh.plotting.figure(
    plot_width=400, 
    plot_height=400,
    x_axis_label='x',
    y_axis_label='y',
    x_range=(-100, 100),
    y_range=(-100, 100),
)

data1 = bokeh.models.sources.ColumnDataSource(data=dict(x=[], y=[]))

circle1 = my_fig.circle(
    "x", "y", 
    source=data1,
    color=palette[0],
    alpha=0.2,
)

handle = bokeh.io.show(my_fig, notebook_handle=True)

new_data1=dict(x=[], y=[])

step = 0
max_step = 1000  # arbitrary stop point for example

while step < max_step:
    new_data1['x'] = [np.random.normal(loc=mu_x1, scale=sigma_x1)]
    new_data1['y'] = [np.random.normal(loc=mu_y1, scale=sigma_y1)]
    
    data1.stream(new_data1)
    
    bokeh.io.push_notebook(handle=handle)
    step += 1
```

# Two Distributions
The above code is now modified so that on each update, a point for a second Normal Gaussian Distribution gets added to the plot.

```python
mu_x1, mu_y1 = np.random.randint(-40, 40, 2)
mu_x2, mu_y2 = np.random.randint(-10, 70, 2)
sigma_x1, sigma_y1 = np.random.randint(4, 15, 2)
sigma_x2, sigma_y2 = np.random.randint(1, 25, 2)


my_fig = bokeh.plotting.figure(
    plot_width=400, 
    plot_height=400,
    x_axis_label='x',
    y_axis_label='y',
    #x_range=(-100, 100),
    #y_range=(-100, 100),
)

data1 = bokeh.models.sources.ColumnDataSource(data=dict(x=[], y=[], x2=[], y2=[]))

circle1 = my_fig.circle(
    "x", "y", 
    source=data1,
    color=palette[2],
    alpha=0.2,)   
    
circle2 = my_fig.diamond(
    "x2", "y2", 
    source=data1,
    color=palette[3],
    alpha=0.2,)

handle = bokeh.io.show(my_fig, notebook_handle=True)

new_data1=dict(x=[], y=[], x2=[], y2=[])

step = 0
max_step = 1000  # arbitrary stop point for example

while step < max_step:
    new_data1['x'] = [np.random.normal(loc=mu_x1, scale=sigma_x1)]
    new_data1['y'] = [np.random.normal(loc=mu_y1, scale=sigma_y1)]
    
    new_data1['x2'] = [np.random.normal(loc=mu_x2, scale=sigma_x2)]
    new_data1['y2'] = [np.random.normal(loc=mu_y2, scale=sigma_y2)]
    
    data1.stream(new_data1)
    
    bokeh.io.push_notebook(handle=handle)
    step += 1
```
